```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Growth Stages
stages = ["Stage 1\nVegetative\n",
          "Stage 2\nReproductive\n",
          "Stage 3\nRipening\n",
]

# RMSE values for XGB and GAT-LSTM
rmse_values_xgb = [0.48, 0.51,0.50]
rmse_values_gat = [0.45,0.46,0.43]

# R values for XGB and GAT-LSTM
r2_scores_xgb = [0.76, 0.73, 0.75]
r2_scores_gat = [0.78, 0.80, 0.79]

# Define color palettes
rmse_colors = sns.color_palette("Blues", len(stages))
r2_colors = sns.color_palette("Greens", len(stages))

# Bar width
bar_width = 0.35  # To place bars side by side

# Create figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 6))  # Two side-by-side plots

# RMSE Bar Plot with XGB vs GAT-LSTM
x = np.arange(len(stages))  # X-axis locations
axes[0].bar(x - bar_width/2, rmse_values_xgb, color=rmse_colors, edgecolor='black', linewidth=1.2, width=bar_width, label="XGB", zorder=2)
axes[0].bar(x + bar_width/2, rmse_values_gat, color=rmse_colors[::-1], edgecolor='black', linewidth=1.2, width=bar_width,hatch="//", label="GAT-LSTM", zorder=2)
axes[0].set_title("RMSE Across Growth Stages (XGB vs GAT-LSTM)", fontsize=16)
axes[0].set_ylabel("Root Mean Square Error (RMSE)", fontsize=14)
axes[0].set_ylim(0, 1)  # Setting y-axis limit for RMSE
axes[0].set_xticks(x)
axes[0].set_xticklabels(stages)
axes[0].grid(axis='y', linestyle='--', alpha=0.6, zorder=1)  # Grid behind bars
axes[0].legend(fontsize=12)

# Display RMSE values on bars
for i in range(len(stages)):
    axes[0].text(i - bar_width/2, rmse_values_xgb[i] + 0.02, f'{rmse_values_xgb[i]}', ha='center', fontsize=12, color='black')
    axes[0].text(i + bar_width/2, rmse_values_gat[i] + 0.02, f'{rmse_values_gat[i]}', ha='center', fontsize=12, color='black')

# R Score Bar Plot with XGB vs GAT-LSTM
axes[1].bar(x - bar_width/2, r2_scores_xgb, color=r2_colors, edgecolor='black', linewidth=1.2, width=bar_width, label="XGB", zorder=2)
axes[1].bar(x + bar_width/2, r2_scores_gat, color=r2_colors[::-1], edgecolor='black', linewidth=1.2, width=bar_width,hatch="//", label="GAT-LSTM", zorder=2)
axes[1].set_title("R Score Across Growth Stages (XGB vs GAT-LSTM)", fontsize=16)
axes[1].set_ylabel("R Score (Higher is better)", fontsize=14)
axes[1].set_ylim(0, 1)  # Ensuring R stays within range
axes[1].set_xticks(x)
axes[1].set_xticklabels(stages)
axes[1].grid(axis='y', linestyle='--', alpha=0.6, zorder=1)  # Grid behind bars
axes[1].legend(fontsize=12)

# Display R values on bars
for i in range(len(stages)):
    axes[1].text(i - bar_width/2, r2_scores_xgb[i] + 0.02, f'{r2_scores_xgb[i]:.2f}', ha='center', fontsize=12, color='black')
    axes[1].text(i + bar_width/2, r2_scores_gat[i] + 0.02, f'{r2_scores_gat[i]:.2f}', ha='center', fontsize=12, color='black')

# Improve layout
plt.tight_layout()
plt.show()
```

```python
from graphviz import Digraph

# Create a new directed graph
dot = Digraph(comment='GTLSTMYieldPrediction Model (No Centroid)', format='png')
dot.attr(rankdir='TB', size='15,15')  # Top-to-bottom layout

# Define node styles
dot.attr('node', shape='box', style='filled', fillcolor='lightblue', fontname='Arial', fontsize='12')
dot.attr('edge', fontname='Arial', fontsize='10')

# Input nodes
dot.node('climate_data', f'Climate Data\n[292, 4, 9]', fillcolor='lightgreen')
dot.node('rs_data', f'Remote Sensing Data\n[292, 4, 4]', fillcolor='lightgreen')

# Climate LSTM
dot.node('climate_lstm', f'Climate LSTM\nLSTM(9, 512, layers=4)\n[292, 512] (last timestep)')
dot.edge('climate_data', 'climate_lstm')

# RS LSTM
dot.node('rs_lstm', f'RS LSTM\nLSTM(4, 512, layers=4)\n[292, 512] (last timestep)')
dot.edge('rs_data', 'rs_lstm')

# Concatenation
dot.node('concat', f'Concatenate\n[292, 1024]', fillcolor='lightyellow')
dot.edge('climate_lstm', 'concat', label='climate_emb')
dot.edge('rs_lstm', 'concat', label='rs_emb')

# Dense Layer
dot.node('dense', f'Dense\nLinear(1024, 32) -> ReLU -> LayerNorm(32)\n[292, 32]')
dot.edge('concat', 'dense')

# GAT Processing
dot.node('gat1', f'GATConv1\n(in=32, out=16, heads=2) -> ReLU\n[292, 32]')
dot.node('gat2', f'GATConv2\n(in=32, out=32, heads=1) -> ReLU -> LayerNorm\n[292, 32]')
dot.edge('dense', 'gat1')
dot.edge('gat1', 'gat2')

# Edge Weights (Adjacency Matrix)
dot.node('edge_weights', f'Edge Weights\n[292, 292]\nSigmoid -> Threshold > 0.5', fillcolor='lightcoral')
dot.edge('edge_weights', 'gat1', label='edge_index')
dot.edge('edge_weights', 'gat2', label='edge_index')

# Residual Connection
dot.node('residual', f'Residual\nScale dense_out\n[292, 32]', fillcolor='lightyellow')
dot.node('combine', f'Add\n[292, 32]', fillcolor='lightyellow')
dot.edge('dense', 'residual')
dot.edge('residual', 'combine')
dot.edge('gat2', 'combine')

# Output Layer
dot.node('yield_mlp', f'Yield MLP\nLinear(32, 32) -> ReLU -> Linear(32, 1)\n[292, 1]')
dot.edge('combine', 'yield_mlp')

# Final Output
dot.node('output', f'Output\nDistrict Yields\n[292, 1]', fillcolor='lightgreen')
dot.edge('yield_mlp', 'output')

# Add Trend (during training/testing)
dot.node('trend_batch', f'Trend Batch\n[292, 1]', fillcolor='lightgreen')
dot.node('final_output', f'Final Output\nPredicted Yields\n[292, 1]', fillcolor='lightgreen')
dot.edge('trend_batch', 'final_output', label='Add')
dot.edge('output', 'final_output')

# Render the diagram
dot.render('GTLSTMYieldPrediction_Model_NoCentroid', view=True)
print("Diagram saved as GTLSTMYieldPrediction_Model_NoCentroid.png")
```

```python
# Install necessary libraries if you haven't already in this Colab session
# !pip install geopandas matplotlib contextily

import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as cx

# List of states to highlight with different colors
states_to_highlight = [
    'Telangana', 'Gujarat', 'Maharashtra', 'Uttar Pradesh', 'Haryana',
    'Punjab', 'Andhra Pradesh', 'Odisha', 'Madhya Pradesh',
    'Karnataka', 'West Bengal', 'Rajasthan', 'Chhattisgarh',
    'Himachal Pradesh', 'Tamil Nadu', 'Bihar', 'Kerala', 'Uttarakhand' # Corrected 'Uttaranchal' to 'Uttarakhand'
]

# URL of a GeoJSON file with Indian states boundaries
# This URL is from a reliable source found in search results, but can change.
url = "https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson"

try:
    # Read the GeoJSON file into a GeoDataFrame
    india_states = gpd.read_file(url)
    print("India states data loaded.")

    # Separate the states into those to highlight and the rest
    # Assuming the state name column is 'ST_NM' - check india_states.head() if needed
    states_highlighted = india_states[india_states['ST_NM'].isin(states_to_highlight)]
    rest_of_india = india_states[~india_states['ST_NM'].isin(states_to_highlight)]

    # Check if filtering was successful for highlighted states
    if states_highlighted.empty:
        print("None of the specified states were found in the data.")
        plot_available = False
    else:
        print(f"{len(states_highlighted)} of the specified states found and filtered.")
        plot_available = True

    if plot_available:
        # Create a figure and axes for the plot
        fig, ax = plt.subplots(1, 1, figsize=(14, 12))

        # --- Plotting ---

        # Reproject data to Web Mercator for basemap compatibility
        # Only reproject if data is not empty
        if not rest_of_india.empty:
             rest_of_india_wm = rest_of_india.to_crs(epsg=3857)
             # Plot the rest of India with only borders
             rest_of_india_wm.plot(ax=ax, facecolor='none', edgecolor='black', linewidth=0.5)
             print("Rest of India states plotted with borders.")
        else:
             print("Rest of India data is empty, cannot plot borders.")


        if not states_highlighted.empty:
            states_highlighted_wm = states_highlighted.to_crs(epsg=3857)
            # Plot the highlighted states with different colors based on state name
            # Using a categorical colormap (e.g., 'tab20') as we have distinct states
            states_highlighted_wm.plot(ax=ax, column='ST_NM', cmap='tab20', legend=True,
                                       legend_kwds={'loc': 'lower right', 'title': 'Highlighted States'},
                                       edgecolor='black', linewidth=0.7)
            print("Highlighted states plotted with different colors.")


        # Add a basemap (optional)
        # Use the CRS of the reprojected data for the basemap
        # Choose a basemap that doesn't obscure the states too much
        if not states_highlighted.empty:
             cx.add_basemap(ax, crs=states_highlighted_wm.crs.to_wkt(), source=cx.providers.CartoDB.Positron, zoom=4) # Adjust zoom level as needed
        elif not rest_of_india.empty:
             cx.add_basemap(ax, crs=rest_of_india_wm.crs.to_wkt(), source=cx.providers.CartoDB.Positron, zoom=4)
        else:
             print("Cannot add basemap as no valid geodataframe is available.")


        # Add a title
        ax.set_title('Specified Indian States Highlighted with Different Colors', fontsize=16)

        # Remove axis for a cleaner map look
        ax.set_axis_off()

        # Display the plot
        plt.show()

except Exception as e:
    print(f"An error occurred during data loading or plotting: {e}")
```
