````python
import pandas as pd

# Load the original dataset
df1 = pd.read_csv('/content/pcp_2003_2017.csv')

# List of unique column indices to drop
cols_to_drop = list(set([185, 230, 251, 374, 505, 706, 738, 57, 377, 26, 29, 33, 351, 352, 375, 376, 421,
                         548, 549, 550, 551, 552, 553, 554, 555, 556, 558, 701, 704, 705, 737, 479, 79,
                         80, 81, 98, 99, 140, 153, 154, 179, 180, 244, 245, 282, 306, 307, 343, 344, 345,
                         589, 590, 743, 744, 745]))

# Drop the columns
df1.drop(df1.columns[cols_to_drop], axis=1, inplace=True)

rows_per_year = 12

filtered_data = []

# Adjusted range to start from 2004
for year in range(2004, 2018):

    # Adjusted indices for 7 months: Oct (row 10) to Apr (row 16) for each year
    # The row indices in the DataFrame are zero-based, so row 10 corresponds to index 9.
    start_index = (year - 2004) * rows_per_year + 9  # 2004 should start at index 9 (row 10)
    end_index = start_index + 7  # Select 7 rows from Oct-Apr (inclusive)

    year_data = df1.iloc[start_index:end_index]  # Select rows for the current year
    filtered_data.append(year_data)

# Combine the filtered data into a single DataFrame
filtered_df1 = pd.concat(filtered_data, ignore_index=True)
# Save the filtered data to CSV
filtered_df1.to_csv('/content/new_df1_pcp7.csv', index=False)
# Display the filtered DataFrame
print(filtered_df1)

# Load the filtered dataset
filtered_df1 = pd.read_csv('/content/new_df1_pcp7.csv')

# Initialize the new DataFrame for all 972 columns
num_years = filtered_df1.shape[0] // 7  # Total years covered
num_columns = filtered_df1.shape[1]  # Number of original columns

# Create the new DataFrame with 7 columns (pcp1 to pcp7) and correct row count
new_df1 = pd.DataFrame({
    f'pcp{i}': np.nan for i in range(1, 8)  # Adjusted for 7 months
}, index=range(num_years * num_columns))  # One row for each year-column block

# Assign values from `filtered_df1` to `new_df1`
row_index = 0  # To track the current row in `new_df1`

for col_index in range(num_columns):
    for year_index in range(num_years):
        start_index = year_index * 7
        end_index = start_index + 7
        if end_index > filtered_df1.shape[0]:
            break

        # Extract 7 months for this column-year block
        year_data = filtered_df1.iloc[start_index:end_index, col_index].values

        # Assign to the corresponding row in new_df1
        #year_data = pd.to_numeric(year_data, errors='coerce')
        new_df1.iloc[row_index, :] = year_data
        row_index += 1

# Save to CSV and verify shape
print(f"new_df1 shape: {new_df1.shape}")  # Should be (49572, 7)
new_df1.fillna(0, inplace=True)
new_df1.to_csv('/content/df1_with_pcp7_values_2004.csv', index=False)

# Display first 10 rows for confirmation
print(new_df1)
````

````python
import pandas as pd
from geopy.distance import great_circle

# Read the CSV files
df_all_grids = pd.read_csv('/content/lon_lat_with_districts_701.csv')
df_yield = pd.read_csv('/content/filtered_district_centroid_701.csv')

# Function to find the 10 nearest points for each district
def find_nearest_points(district, centroid_lon, centroid_lat, df_all_grids, num_points=10):
    distances = []
    origin = (centroid_lat, centroid_lon)

    # Calculate distance from the centroid to all points in df_all_grids
    for _, row in df_all_grids.iterrows():
        point = (row['Latitude'], row['Longitude'])
        distance = great_circle(origin, point).kilometers
        if distance > 0:  # Skip zero distance
            distances.append((distance, row['Longitude'], row['Latitude'], row['DISTRICT']))

    # Sort by distance and select the nearest points
    distances.sort(key=lambda x: x[0])
    nearest_points = distances[:num_points]

    # Prepare results
    results = []
    for dist, lon, lat, nearest_district in nearest_points:
        results.append({
            'District': district,
            'Centroid Longitude': centroid_lon,
            'Centroid Latitude': centroid_lat,
            'Nearest District': nearest_district,
            'Nearest Longitude': lon,
            'Nearest Latitude': lat,
            'Distance (km)': dist
        })

    return results

# Create a DataFrame to store the results
results = []

# Iterate through each district in df_yield
for _, row in df_yield.iterrows():
    district = row['DISTRICT']
    centroid_lon = row['Longitude']
    centroid_lat = row['Latitude']

    # Find the 10 nearest points for the current district
    nearest_points = find_nearest_points(district, centroid_lon, centroid_lat, df_all_grids, num_points=10)

    # Append to results
    results.extend(nearest_points)

# Create a DataFrame from the results
df_results = pd.DataFrame(results)

# Display the results
print(df_results.head())

# Save the results to a CSV file
df_results.to_csv('/content/nearest_points_for_each_district_from_centroid_6s_701.csv', index=False)
````

# Detrending yield

````python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('/content/reshaped_dataset_7M_701grids_with year.csv')

# Create an empty DataFrame to store the detrended yields and trends
detrended_df = pd.DataFrame()

# Function to calculate Sen's slope, detrend, and save the trend
def detrend_sens_slope(df):
    x = np.arange(len(df))  # Time variable (0, 1, 2, ..., 51 for 52 years)
    y = df['Yield'].values

    # Calculate the Sen's slope
    slopes = []
    for i in range(len(y)):
        for j in range(i + 1, len(y)):
            slopes.append((y[j] - y[i]) / (x[j] - x[i]))
    slope = np.median(slopes)

    # Calculate the trend (slope * time)
    df['Trend'] = slope * x

    # Detrend the yield data by subtracting the trend from the original yield
    df['Detrended_Yield'] = df['Yield'] - df['Trend']

    return df

# Apply detrending and trend calculation for each district
for district in df['District'].unique():
    df_district = df[df['District'] == district].copy()
    df_district = detrend_sens_slope(df_district)
    detrended_df = pd.concat([detrended_df, df_district], ignore_index=True)

# Save the updated DataFrame with the detrended yield and trend
detrended_df.to_csv('/content/detrended_district_yearly_yield_6s_1999_128g.csv', index=False)

# Display the first few rows of the updated DataFrame
print(detrended_df.head())
````
